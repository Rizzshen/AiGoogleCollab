```python
def sum_nested_list(nested_list):
  total = 0
  for element in nested_list:
    if isinstance(element, list): # Check if the element is a list
      total += sum_nested_list(element) # Recursively sum the nested list
    else:
      total += element # Add the number to the total
  return total
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list))
```

#Task1

```python
def min_coins(coins, amount):
  dp = [float('inf')] * (amount + 1)
  dp[0] = 0
  for coin in coins:
    for i in range(coin, amount + 1):
      dp[i] = min(dp[i], dp[i - coin] + 1)
  return dp[amount] if dp[amount] != float("inf") else -1
coins=[1,2,5]
print(min_coins(coins, amount=11))
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)

    # Create DP table (n+1) x (m+1) initialized to 0
    dp = [[0] * (m + 1) for _ in range(n + 1)]

    # Fill the table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:          # characters match
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:                               # characters don't match
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]
print(longest_common_subsequence("abcde", "ace"))
```

```python
def knapsack(weights, values, capacity):
    n = len(weights)

    # DP table of size (n+1) x (capacity+1)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                # Option 1: include the item
                include = values[i - 1] + dp[i - 1][w - weights[i - 1]]
                # Option 2: exclude the item
                exclude = dp[i - 1][w]
                dp[i][w] = max(include, exclude)
            else:
                # Cannot include the item
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

weights=[1,3,4,5]
values=[1,4,5,7]
capacity = 7
print(knapsack(weights, values, capacity))
```
