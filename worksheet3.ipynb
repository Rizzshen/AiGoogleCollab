```python
#3.1
#1
import pandas as pd

# Upload the CSV file (you will need to upload titanic.csv in Colab)
from google.colab import files
uploaded = files.upload()

# Load the uploaded file
df = pd.read_csv(list(uploaded.keys())[0])

# Display the first few rows
print("----- Head of Dataset -----")
display(df.head())

print("----- DataFrame Info -----")
df.info()


print("\n----- Shape (Rows, Columns) -----")
print(df.shape)

print("\n----- Missing Values Per Column -----")
print(df.isnull().sum())

print("\n----- Summary Statistics (Numerical Columns) -----")
display(df.describe())

print("\n----- Data Types -----")
print(df.dtypes)
```

```python
#Problem1
fare = df[["Fare"]]
print("----- Fare (head) -----")
display(fare.head())

class_age = df[["Pclass", "Age"]]
print("----- Class & Age (head) -----")
display(class_age.head())

survived_gender = df[["Survived", "Sex"]]
print("----- Survived & Gender (head) -----")
display(survived_gender.head())
```

```python
#Problem 2
fare_gt_100 = df[df["Fare"] > 100]
print("----- Fare > 100 -----")
display(fare_gt_100.head())

first_class = df[df["Pclass"] == 1]
print("----- First Class (Pclass = 1) -----")
display(first_class.head())

female_under_18 = df[(df["Age"] < 18) & (df["Sex"] == "female")]
print("----- Female Under 18 -----")
display(female_under_18.head())

embarked_c_or_s = df[df["Embarked"].isin(["C", "S"])]
print("----- Embarked = C or S -----")
display(embarked_c_or_s.head())

first_second_class = df[df["Pclass"].isin([1, 2])]
print("----- Pclass in [1, 2] -----")
display(first_second_class.head())
```

```python

#3.2
df["fare_per_year"] = df["Fare"] / df["Age"]

df["fare_per_year"] = df["fare_per_year"].replace([float('inf')], None)

high_fare_age = df[df["fare_per_year"] > 5]

high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

result_age = high_fare_age_srt[["Name", "fare_per_year"]]

print("----- Passenger with Highest Fare per Year (Fare/Age) -----")
display(result_age.head())

df["fare_per_class"] = df["Fare"] / df["Pclass"]

adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)]

adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)

result_class = adult_males_srt[["Name", "Age", "fare_per_class"]]

print("----- Adult Male with Highest Fare per Class (Fare/Pclass) -----")
display(result_class.head())
```

```python
#3.3
total_fare = df["Fare"].sum()

fare_class_1 = df[df["Pclass"] == 1]["Fare"].sum()
fare_class_2 = df[df["Pclass"] == 2]["Fare"].sum()
fare_class_3 = df[df["Pclass"] == 3]["Fare"].sum()

fare_totals = [fare_class_1, fare_class_2, fare_class_3]

fare_percent = [(f / total_fare) * 100 for f in fare_totals]

print("===== Percent of Total Fare Revenue by Class =====")
print(f"First Class:  {fare_percent[0]:.2f}%")
print(f"Second Class: {fare_percent[1]:.2f}%")
print(f"Third Class:  {fare_percent[2]:.2f}%\n")

def age_group(age):
    if pd.isnull(age):
        return None
    elif age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

df["age_group"] = df["Age"].apply(age_group)

total_passengers = len(df)

age_counts = df["age_group"].value_counts()

age_percent = (age_counts / total_passengers) * 100

print("===== Percent of Passengers in Each Age Group =====")
display(age_percent)
```
